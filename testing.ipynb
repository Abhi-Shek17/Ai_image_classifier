```python
import torch
from torchvision import datasets,transforms
import random as random
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
import torch
model=torch.load('/kaggle/input/baaprebaap/pytorch/1sttimegwok/2/checkpoint.pth',map_location=torch.device('cpu'))
model
```

```python
transform=transforms.Compose([
    transforms.Resize(size=(512,512)),
    transforms.ToTensor()
])
from pathlib import Path
data_path = '/kaggle/input/ai-cla/'
data=datasets.ImageFolder(root=data_path,transform=transform)
```

```python
class_names=data.classes
class_names
```

```python
tloader = DataLoader(data,
                    batch_size=50,
                    num_workers=4)
```

```python
def viz_tensor(img_single,label):
    i=0
    for img_new in img_single:
        i+=1
        img_new=img_new.permute(1,2,0).to('cpu')
        plt.figure(figsize=(10,7))
        plt.imshow(img_new)
        plt.axis("off")
        plt.title(label)
        if i>10:
            break
```

```python
TP=0
TN=0
FP=0
FN=0
wrong_dataset=[]
for X,y in tloader:
    X,y=X.to(device),y.to(device)
    pred=model(X)
    pred=torch.argmax(pred,dim=1)
    for i,val in enumerate(y):
        if val==1:
            if pred[i]==1:
                TP+=1
            else:
                FP+=1
                wrong_dataset.append(X[i])
        else:
            if pred[i]==0:
                TN+=1
            else:
                FN+=1
                wrong_dataset.append(X[i])
print(f"TP,TN,FP,FN : {TP},{TN},{FP},{FN}")                
```
